```python
import pandas as pd
df = pd.read_csv("PD 2024 Wk 1 Input.csv")

df.head()
```

```python
df [['Date', 'Flight Number', 'From-To', 'Class', 'Price']] = df['Flight Details'].str.split('//', expand = True)
```

```python
df.head()
```

```python
df[['From', 'To']]=df['From-To'].str.split('-', expand = True)
```

```python
df.head()
```

```python
df.drop('Flight Details', axis=1, inplace = True)
```

```python
df.head()
```

```python
order = ['Date', 'Flight Number', 'From', 'To', 'Class', 'Price', 'Flow Card?', 'Bags Checked', 'Meal Type']
df = df[order]
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Price']= pd.to_numeric(df['Price'])
```

```python
df.head()
```

```python
df.loc[df['Flow Card?']== 1, 'Flow Card?'] = 'Yes'
```

```python
df.head()
```

```python
df.loc[df['Flow Card?'] == 0, 'Flow Card?'] = 'No'
```

```python
df.head()
```

```python
flowCardHolders = df[df['Flow Card?'] == 'Yes']
```

```python
nonFlowCardHolders = df[df['Flow Card?'] == 'No']
```

```python
flowCardHolders.to_csv('Flow Card Holders.csv', index=False)
nonFlowCardHolders.to_csv('Non-Flow Card Holders.csv', index=False)
```
