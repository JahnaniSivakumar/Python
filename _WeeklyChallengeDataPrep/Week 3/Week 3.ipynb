```python
import pandas as pd
import numpy as np
```

```python
flowCard = pd.read_csv("PD 2024 Wk 1 Output Flow Card.csv")
nonFlowCard = pd.read_csv("PD 2024 Wk 1 Output Non-Flow Card.csv")
t1 = pd.read_excel("PD 2024 Wk 3 Input.xlsx", sheet_name = 'Q1')
t2 = pd.read_excel("PD 2024 Wk 3 Input.xlsx", sheet_name = 'Q2')
t3 = pd.read_excel("PD 2024 Wk 3 Input.xlsx", sheet_name = 'Q3')
t4 = pd.read_excel("PD 2024 Wk 3 Input.xlsx", sheet_name = 'Q4')
```

```python
sales = pd.concat([flowCard,nonFlowCard], ignore_index=True)
target = pd.concat([t1,t2,t3,t4], ignore_index=True)
```

```python

sales['Class'].replace({'Economy': 'First Class', 'First Class': 'Economy', 'Business Class' : 'Premium Economy', 'Premium Economy': 'Business Class'}, inplace = True)
```

```python
sales.head()
```

```python
sales['Class'] = sales['Class'].str[0]  # Use first letter of the string if it's a single word
```

```python
sales.head()
```

```python


sales['Date'] = pd.to_datetime(sales['Date'])
sales['Price'] = pd.to_numeric(sales['Price'])
sales['Date'] = sales['Date'].dt.month

sales.head(50)
```

```python
sales_sum = sales.groupby(['Class', 'Date'])['Price'].sum().reset_index()
```

```python
sales_sum.head(50)
```

```python


target['Class']= target['Class'].str[0]
target.head(50)
```

```python
df = pd.merge(sales_sum, target, left_on = ['Class', 'Date'], right_on = ['Class', 'Month']) 
```

```python
df.head(50)
```

```python
df['Difference to Target'] = df['Price'] - df['Target']
```

```python
df.head()
```

```python
df.shape

order = ['Difference to Target', 'Month', 'Price', 'Class', 'Target']
df = df[order]
```

```python
df.to_csv("Week3Output.csv", index=False)
```
